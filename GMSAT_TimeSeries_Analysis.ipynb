```python
import pandas as pd                          
import matplotlib.pyplot as plt 
import numpy as np
import cmocean 

%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6   
```

```python
#Reading in avg global surface data
gmt_Berkeley_Earth = pd.read_csv('./data/temp/gmt_Berkeley_Earth.csv')
gmt_ERA5 = pd.read_csv('./data/temp/gmt_ERA5.csv')
gmt_GISTEMP =pd.read_csv('./data/temp/gmt_GISTEMP.csv')
gmt_HadCrut5 = pd.read_csv('./data/temp/gmt_HadCRUT5.csv')
gmt_JRA55 =pd.read_csv('./data/temp/gmt_JRA-55.csv')
gmt_NOAA_Global_Temp =pd.read_csv('./data/temp/gmt_NOAAGlobalTemp.csv')
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

ax.plot(gmt_Berkeley_Earth['Year'],
        gmt_Berkeley_Earth['Berkeley Earth (degC)'],
        linewidth=1,label='Berkeley Earth Global Mean Temp Data',
        color='red')

#legend
ax.legend() 

# title
ax.set_title('Berkeley Earth Average Temperature Anomoly Data in Degrees Celsius')

#line at zero
ax.axhline(0,color='black')

#background color
ax.set_facecolor("lightblue")
ax.axhline(0,color='white',linewidth=0.5)

# axes labels
ax.set_ylabel('Mean Temp Anomoly (deg C)')
ax.set_xlabel('Year')
```

Figure 1. Average surface temperature anomoly data provided by Berkeley Earth from 1850-present. This graph shows an increasing trend and indicates a rising average temperature.

```python
fig, ax = plt.subplots(figsize = (10, 5))

ax.plot(gmt_Berkeley_Earth['Year'],
        gmt_Berkeley_Earth['Berkeley Earth (degC)'],
        linewidth=2,label='Berkeley Earth Global Mean Temp Data',
        color='black')

ax.plot(gmt_ERA5['Year'],
        gmt_ERA5['ERA5 (degC)'],
        linewidth=2,label='ERA5 Mean Temp Data',
        color='red')

ax.plot(gmt_GISTEMP['Year'],
        gmt_GISTEMP['GISTEMP (degC)'],
        linewidth=2,label='GISTEMP Mean Temp Data',
        color='blue')

ax.plot(gmt_HadCrut5['Year'],
        gmt_HadCrut5['HadCRUT5 (degC)'],
        linewidth=2,label='HadCRUT5 Mean Temp Data',
        color='yellow')

ax.plot(gmt_JRA55['Year'],
        gmt_JRA55['JRA-55 (degC)'],
        linewidth=2,label='JRA55 Mean Temp Data',
        color='green')

ax.plot(gmt_NOAA_Global_Temp['Year'],
        gmt_NOAA_Global_Temp['NOAAGlobalTemp (degC)'],
        linewidth=2,label='NOAAGlobal Mean Temp Data',
        color='purple')



ax.legend() 

ax.set_title('Average Temperature Anomoly Data From Six Sources in Degrees Celsius')

ax.set_ylabel('Mean Temp Anomoly (deg C)')
ax.set_xlabel('Year')

```

Figure 7. Global average temperature anomoly data from Berkeley Earth, ERA5, GISTEMP, HadCRUT5, JRA55, and NOAAGlobalTemp. Time period varies by source, but is generally between 1850 and present day. Trend is consistent for all sources, even though there is some slight variability, showing an incline and a rising average surface temperature.
